###  PCA를 적용한 고차원 군집의 시각화

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.decomposition import PCA
%matplotlib inline
```

#### 1. 시뮬레이션으로 데이터 만들고 시각화 해보기.

```python
# 8차원 공간에서 3개의 군집 생성.
# 시각화가 어렵다.
X, label = make_blobs(n_samples=300, n_features=8, centers=3, cluster_std = 4, random_state=123)
X.shape
```

```python
# 컬러 팔레트 정의.
myColors = {0:'red',1:'green',2:'blue'}
#labelColors = pd.Series(label).apply(lambda x: myColors[x])
labelColors=[myColors[x] for x in label]
```

```python
# 원 좌표축 중에서 2개씩 가져다가 평면을 정의하고 시각화 해본다. (X0 vs X1)
# 군집이 서로 겹치고 구별하기 어렵다.
plt.scatter(X[:,0],X[:,1],marker="o",alpha=0.7, s=10, c=labelColors)
plt.xlabel("X0")
plt.ylabel("X1")
plt.show()
```

```python
# 원 좌표축 중에서 2개씩 가져다가 평면을 정의하고 시각화 해본다. (X0 vs X4)
# 군집이 서로 겹치고 구별하기 어렵다.
plt.scatter(X[:,0],X[:,4],marker="o",alpha=0.7, s=10, c=labelColors)
plt.xlabel("X0")
plt.ylabel("X4")
plt.show()
```

```python
# 원 좌표축 중에서 2개씩 가져다가 평면을 정의하고 시각화 해본다. (X4 vs X7)
# 군집이 서로 겹치고 구별하기 어렵다.
plt.scatter(X[:,4],X[:,7],marker="o",alpha=0.7, s=10, c=labelColors)
plt.xlabel("X4")
plt.ylabel("X7")
plt.show()
```

#### 2. PC1과 PC2로 정의된 2D 평면에 투영하여 시각화 하기.

```python
# PC1과 PC2만 필요하다.
pca = PCA(n_components = 2)
X_pca = pca.fit_transform(X)
```

```python
# PC1과 PC2로 정의된 2D 평면에 투영.
# 군집의 겹침은 최소!!!
plt.scatter(X_pca[:,0],X_pca[:,1],marker="o",alpha=0.7, s=10, c=labelColors)
plt.xlabel("PC1")
plt.ylabel("PC2")
plt.show()
```
