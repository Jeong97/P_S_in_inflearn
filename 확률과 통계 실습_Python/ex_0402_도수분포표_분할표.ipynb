### 도수분포표와 분할표

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import io
from google.colab.files import upload
pd.set_option('display.precision',3)
```

#### 1. 데이터셋을 읽어 온다.

```python
dict_data = upload()
```

```python
df = pd.read_csv(io.BytesIO(dict_data['data_studentlist_en.csv']))    # 파일 읽어들이기.
```

```python
df.head(5)
```

#### 2. 도수분포표.

```python
bloodType = df['bloodtype']
gender = df['gender']
```

```python
bloodType.value_counts()
```

```python
gender.value_counts()
```

```python
bloodType.value_counts()/bloodType.size   # 또는 bloodType.value_counts(normalize=True)
```

```python
gender.value_counts()/gender.size    # 또는 gender.value_counts(normalize=True)
```

```python
# 도수표 시각화.
bloodType.value_counts().plot.bar(color=['blue','orange','green','red'])
plt.show()
```

```python
# 상대도수표 시각화 (막대).
(bloodType.value_counts()/bloodType.size).plot.bar(color=['blue','orange','green','red'])
plt.show()
```

```python
# 상대도수표 시각화 (pie).
bloodType.value_counts().plot.pie()
plt.show()
```

```python
# 도수표 시각화.
sns.countplot(data=df, x='gender')
plt.show()
```

```python
# 도수표 시각화.
sns.countplot(data=df,x='bloodtype')
plt.show()
```

#### 3. 분할표.

```python
pd.crosstab(bloodType, gender)
```

```python
pd.crosstab(bloodType,gender,margins=True)
```

```python
# 분할표 시각화.
pd.crosstab(bloodType,gender).plot.bar()
plt.show()
```

```python
# 시각화.
pd.crosstab(bloodType,gender).plot.bar(stacked=True, color=["red","blue"])
plt.legend(loc="upper left")
plt.show()
```

```python
# 시각화.
pd.crosstab(gender,bloodType).plot.bar()
plt.show()
```

```python
# 시각화.
pd.crosstab(gender,bloodType).plot.bar(stacked=True)
plt.legend(loc="upper left")
plt.show()
```

```python
# 시각화.
sns.countplot(data=df, x='bloodtype', hue='gender')
plt.legend(loc="upper left")
plt.show()
```
