### 상관계수 계산

```python
import pandas as pd
import numpy as np
import scipy.stats as st
import matplotlib.pyplot as plt
import seaborn as sns
```

#### 1. 데이터셋을 읽어 와서 탐색한다.

```python
# Seaborn 데이터셋 관련해서는 다음 링크를 참고해 본다:
# https://github.com/mwaskom/seaborn-data
df = sns.load_dataset('iris')         # Seaborn 라이브러리의 내장 데이터. iris = 붓꽃 데이터.
```

```python
df.shape
```

```python
df.head(3)
```

```python
x = df['petal_length']
y = df['sepal_length']
```

```python
plt.scatter(x, y, color= 'green', alpha=0.5, marker ='o', s=30 )
plt.xlabel('Petal Length')
plt.ylabel('Sepal Length')
plt.title('IRIS FLOWER SCATTER PLOT')
plt.show()
```

#### 2. Pearson 상관계수.

$H_0$: 상관계수는 0과 다르지 않다.   <br>
$H_1$: 상관계수는 0과 유의하게 다르다.

```python
np.round(st.pearsonr(x,y),3)           # 상관계수와 p-값.
```

Pandas 사용.

```python
x.corr(y)
```

```python
np.round(df.drop(columns=['species']).corr(),3)                  # species 컬럼은 제외하고 나머지 수치형 컬럼으로 상관계수 행렬 계산.
```

#### 3. Spearman 상관계수.

```python
np.round(st.spearmanr(x,y),3)          # 상관계수와 p-값.
```

#### 4. Kendall 상관계수.

```python
np.round(st.kendalltau(x,y),3)         # 상관계수와 p-값.
```
