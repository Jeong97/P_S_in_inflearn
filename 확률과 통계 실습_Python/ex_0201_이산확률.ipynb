### 이산확률분포

```python
import scipy.stats as st
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

#### 1. 이항확률분포.

```python
n=10
p=0.5
```

*확률분포*

```python
st.binom.pmf(5, n, p)               
```

*분위수*

```python
st.binom.ppf(0.3, n, p)             
```

*누적확률 (구간의 확률)*

```python
# P(0 <= x <= 5)
st.binom.cdf(5,n,p)
```

```python
# P(3 <= x <=7)
st.binom.cdf(7,n,p)-st.binom.cdf(2,n,p)        
```

*시각화*

```python
# 확률분포.
x=np.arange(0,11)
plt.scatter(x, st.binom.pmf(x,n,p),color='green')
plt.show()
```

```python
# 분위수.
a=np.arange(0,11)/10.0
plt.scatter(a, st.binom.ppf(a,n,p),color='red')
plt.show()
```

#### 2. 푸아송확률분포.

```python
lamb = 5
```

*확률분포*

```python
st.poisson.pmf(2,lamb)                    
```

*분위수*

```python
st.poisson.ppf(0.5,lamb)                 
```

*누적확률 (구간의 확률)*

```python
# P(0 <= x <= 5)
st.poisson.cdf(5,lamb)
```

```python
# P(3 <= x <=7)
st.poisson.cdf(7,lamb)-st.poisson.cdf(2,lamb)
```

*시각화*

```python
# 확률분포.
x=np.arange(0,11)
plt.scatter(x, st.poisson.pmf(x,lamb),color='green')
plt.show()
```

```python
# 분위수.
a=np.arange(0,11)/10.0
plt.scatter(a, st.poisson.ppf(a,lamb),color='red')
plt.show()
```
