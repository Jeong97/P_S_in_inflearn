### 로지스틱 회귀를 사용한 분류 예측

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn import metrics
%matplotlib inline
```

#### 1. 데이터를 읽어오고 탐색한다.

```python
# 내장 데이터 읽어오기.
data = load_breast_cancer()
```

```python
# 데이터에 대한 설명 보기.
print(data.DESCR)
```

```python
# 설명 변수.
X = data['data']
print(data['feature_names'])
```

```python
X.shape
```

```python
# 반응 변수.
# 0 = 'benign'(양성).
# 1 = 'malignant' (악성).
Y = 1 - data['target']
label = list(data['target_names'])
label.reverse()
print(label)
```

```python
# 도수표 시삭화.
ser = pd.Series(Y)
table = ser.value_counts()
table = table.sort_index()                                         # 인덱스에 소팅 적용.
plt.bar(label,table.values)
plt.show()
```

#### 2. 트레이닝과 테스팅.

```python
# 데이터를 Training과 Testing 용으로 쪼갠다.
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.4, random_state=1234)
```

```python
print(X_train.shape)
print(X_test.shape)
print(Y_train.shape)
print(Y_test.shape)
```

```python
# 학습과 예측.
LR = LogisticRegression(max_iter=1000,solver='newton-cg')
LR.fit(X_train,Y_train)
Y_pred_test = LR.predict(X_test)        # Out-of-sample prediction. Cutoff = 0.5 fixed.
```

```python
# 혼동 행렬.
conf_mat = metrics.confusion_matrix(Y_test, Y_pred_test)
print(conf_mat)
```

```python
# 혼동행렬을 사용한 Accuracy, Sensitivity, Specificity, Precision 계산.
accuracy = (conf_mat[0,0] + conf_mat[1,1])/np.sum(conf_mat)
sensitivity = conf_mat[1,1]/(conf_mat[1,0]+conf_mat[1,1])
specificity = conf_mat[0,0]/(conf_mat[0,0]+conf_mat[0,1])
precision = conf_mat[1,1]/(conf_mat[0,1]+conf_mat[1,1])
print('Accuracy    = {}'.format(np.round(accuracy,3)))
print('Sensitvity  = {}'.format(np.round(sensitivity,3)))
print('Specificity = {}'.format(np.round(specificity,3)))
print('Precision   = {}'.format(np.round(precision,3)))
```

```python
# 다른 방법.
accuracy = metrics.accuracy_score(Y_test,Y_pred_test)
sensitivity = metrics.recall_score(Y_test,Y_pred_test)
precision = metrics.precision_score(Y_test,Y_pred_test)
print('Accuracy    = {}'.format(np.round(accuracy,3)))
print('Sensitvity  = {}'.format(np.round(sensitivity,3)))
print('Precision   = {}'.format(np.round(precision,3)))
```

#### 3. Cutoff (기준 확률).

```python
# Y = 1일 확률 (조건부 확률) 예측.
Y_pred_test_prob=LR.predict_proba(X_test)[:,1]
```

```python
# Cutoff를 원하는 값으로,
cutoff = 0.2                                                    # Cutoff는 0과 1 사이의 수치.
Y_pred_test_val = (Y_pred_test_prob > cutoff).astype(int)
conf_mat = metrics.confusion_matrix(Y_test,Y_pred_test_val)
print(conf_mat)
```

```python
accuracy = (conf_mat[0,0] + conf_mat[1,1])/np.sum(conf_mat)
sensitivity = conf_mat[1,1]/(conf_mat[1,0]+conf_mat[1,1])
specificity = conf_mat[0,0]/(conf_mat[0,0]+conf_mat[0,1])
precision = conf_mat[1,1]/(conf_mat[0,1]+conf_mat[1,1])
print('Accuracy    = {}'.format(np.round(accuracy,3)))
print('Sensitivity  = {}'.format(np.round(sensitivity,3)))
print('Specificity = {}'.format(np.round(specificity,3)))
print('Precision   = {}'.format(np.round(precision,3)))
```

#### 4. ROC 곡선.

```python
# 초기화.
cutoff_grid = np.linspace(0.0,1.0,101)
TPR = []                                                   # True Positive Rate.
FPR = []                                                   # False Positive Rate.
```

```python
# TPR과 FPR 리스트.
for cutoff in cutoff_grid:
    Y_pred_test_val = (Y_pred_test_prob > cutoff).astype(int)
    conf_mat = metrics.confusion_matrix(Y_test,Y_pred_test_val)
    sensitivity = conf_mat[1,1]/(conf_mat[1,0]+conf_mat[1,1])
    specificity = conf_mat[0,0]/(conf_mat[0,0]+conf_mat[0,1])
    TPR.append(sensitivity)
    FPR.append(1-specificity)
```

```python
# 시각화.
plt.plot(FPR,TPR,c='red',linewidth=1.0)
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curve')
plt.show()
```

#### 5. ROC 곡선 (Sklearn).

```python
# Scikit-Learn 함수를 사용해서 FPR, TPR 계산.
FPR, TPR, cutoffs = metrics.roc_curve(Y_test,Y_pred_test_prob,pos_label=1)      # positive label = 1.
```

```python
# 시각화.
plt.plot(FPR,TPR,c='red',linewidth=1.0)
plt.xlabel('False Positive')
plt.ylabel('True Positive')
plt.title('ROC Curve')
plt.show()
```

```python
# AUC.
auc = metrics.roc_auc_score(Y_test,Y_pred_test_prob)
print('AUC  = {}'.format(np.round(auc,3)))
```
