```python
import numpy as np
import scipy.stats as st
import matplotlib.pyplot as plt
```

양측검정

```python
mu_0 = 0
st.ttest_1samp(x, mu_0)
```

```python
# 검정통계량 수식을 사용해서 계산.
mu_0 = 0
n = len(x)
x_mean = x.mean()
SEM = x.std(ddof=1)/np.sqrt(n)
(x_mean-mu_0)/SEM
```

#### 2. 모집단 2개.

```python
# mu = 3, sigma = 2, n = 10 표본 시뮬레이션.
x1 = np.random.randn(10)*2+3
# mu = 5, sigma = 3, n = 20 표본 시뮬레이션.
x2 = np.random.randn(20)*3+5
n1 = len(x1)
n2 = len(x2)
x1_mean = x1.mean()
x2_mean = x2.mean()
s1 = x1.std(ddof=1)
s2 = x2.std(ddof=1)
```

양측검정 (독립)

```python
st.ttest_ind(x1,x2,equal_var=True)                                        
```

```python
st.ttest_ind_from_stats(x1_mean,s1,n1,x2_mean,s2,n2,equal_var=True)       
```

```python
# 바틀렛 테스트. H0 = 합동분산. H1 = 이분산.
st.bartlett(x1,x2)                  
```

양측검정 (대응)

```python
x1 = np.random.randn(10)*5 + 3
x2 = x1 + np.random.randn(10)*1 + 3
st.ttest_rel(x1,x2)
```
