### 이론적 분위수 (모집단)

```python
import scipy.stats as st
import numpy as np
```

#### 1. 표준정규분포.

주요 분위수를 계산해 본다.

```python
mu = 0
sigma = 1
a = 0.95
st.norm.ppf(a, loc = mu , scale = sigma)
```

```python
a = 0.975
st.norm.ppf(a, loc = mu , scale = sigma)
```

#### 2. Student t 분포.

주요 분위수를 계산해 본다.

```python
# 자유도가 커질수록 표준정규분포의 분위수에 근접한다.
a = 0.95
for df in [10,50,100,2000,100000]:              # df = degree of freedom. 자유도.
    val = st.t.ppf(a, df)
    print('df = %7d,  분위수 = %.5f' % (df,val))
```

```python
# 자유도가 커질수록 표준정규분포의 분위수에 근접한다.
a = 0.975
for df in [10,50,100,2000,100000]:
    val = st.t.ppf(a, df)
    print('df = %7d,  분위수 = %.5f' %(df,val))
```

#### 3. 카이제곱 분포.

다음과 같이 누적확률과 분위수는 서로 역함수 관계임을 확인해 본다.

자유도 5의 카이제곱분포를 따르는 확률변수 X가 있다. 누적확률 P(0<= X <= 8)를 계산해 본다. 

```python
df = 5
print( '%.4f' % st.chi2.cdf(8, df) )
```

자유도 5의 카이제곱분포를 따르는 확률변수 X가 있다. 누적 확률이 0.8438이라면 X의 값은?

```python
df = 5
print( '%.2f' % st.chi2.ppf(0.8438, df) )
```
